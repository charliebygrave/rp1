```python
# initialisation helper functions

def initialise_random(model_initialisation):
    # convert to dict
    initialisations = initialisation_to_dict(model_initialisation)               
    nodes = initialisations.keys()
    
    # set values to 'Random'
    initialisations = {}
    for node in nodes:
        initialisations[node] = 'Random'
    
    # convert to initialisation
    model_initialisation = dict_to_initialisation(initialisations)
    return model_initialisation
        

def initialisation_to_dict(model_initialisation):
    initialisation_lines = []
    for line in model_initialisation.split('\n'):
        if not (line.startswith('#') or line == ''):
            initialisation_lines.append(line.split()[0])

    initialisation_dict = {}
    for line in initialisation_lines:
        if '=' in line:
            key, value =  line.split('=')
            initialisation_dict[key.strip()] = value.strip()
    return initialisation_dict

def dict_to_initialisation(initialisation_dict):
    model_initialisation="# initialise nodes\n"
    for (node, initialisation) in dict.items(initialisation_dict):
         model_initialisation+=node+'='+initialisation+'\n'
    return model_initialisation
    
```
