```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('network-lysosome-biogenesis-6.csv')
```

```python
edgelist = df[['node1', 'node2']].copy()
edgelist.node1 = edgelist.node1.str.replace(' ', '_')
edgelist.node2 = edgelist.node2.str.replace(' ', '_')

edgelist = edgelist.values.tolist()
```

```python
f = open('edgelist-lysosome-biogenesis-1', 'w')
for edge in edgelist:
    f.write('{0}\t{1}\n'.format(edge[0], edge[1]))
f.close()
```

```python
f = open('edgelist-lysosome-biogenesis-1', 'r')
for line in f:
    print(line, end="")
f.close()
```

```python
G = nx.read_edgelist('edgelist-lysosome-biogenesis-1')
```

```python
plt.figure(figsize=(24,24))
nx.draw(G, node_size=50, with_labels= True)
```
